# More examples

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import xarray as xr
import xesmf as xe
```

## Curvilinear grid with missing value

```python
ds = xr.tutorial.load_dataset('rasm')
ds
```

```python
plt.figure(figsize=(12,2));
ax = plt.axes(projection=ccrs.PlateCarree());
ds['Tair'][0].plot.pcolormesh(ax=ax, x='xc', y='yc');
ax.coastlines();
```

```python
ds.rename({'xc': 'lon', 'yc': 'lat'}, inplace=True)
ds
```

```python
ds_out = xe.util.grid_global(5, 4)
```

```python
regridder = xe.Regridder(ds, ds_out, 'bilinear')
```

```python
dr_out = regridder(ds['Tair'])
```

```python
dr_out
```

```python
plt.figure(figsize=(8,4));
ax = plt.axes(projection=ccrs.PlateCarree());
dr_out[0].plot.pcolormesh(ax=ax, x='lon', y='lat');
ax.coastlines();
```

## Coordinate as 1D array

```python
ds = xr.tutorial.load_dataset('air_temperature')
ds
```

```python
ds['lon'].max(), ds['lon'].min(), ds['lat'].max(), ds['lat'].min()
```

```python
dr = ds['air']
ax = plt.axes(projection=ccrs.PlateCarree())
dr.isel(time=0).plot.pcolormesh(ax=ax, vmin=230, vmax=300);
ax.coastlines();
```

```python
ds_out = {'lat': np.linspace(20, 60, 100),
          'lon': np.linspace(220, 300, 200)
         }
```

```python
regridder = xe.Regridder(ds, ds_out, 'bilinear')
regridder
```

```python
dr_out = regridder(dr)
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
dr_out.isel(time=0).plot.pcolormesh(ax=ax, vmin=230, vmax=300);
ax.coastlines();
```

```python
ds_global = xe.util.grid_global(5, 4)
```

```python
regridder_to_global = xe.Regridder(ds, ds_global, 'bilinear')
regridder_to_global
```

```python
dr_out_global = regridder_to_global(dr)
```

```python
dr_out_global
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
dr_out_global.isel(time=0).plot.pcolormesh(ax=ax, vmin=230, vmax=300,
                                           x='lon', y='lat');
ax.coastlines();
```

## Dictionary as 1D grid

```python
regridder_from_dict = xe.Regridder(ds, 
                                   {'lon': np.linspace(-180,180), 
                                    'lat': np.linspace(-90,90)},
                                   'bilinear')
regridder_from_dict
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
(regridder_from_dict(dr)
.isel(time=0)
.plot
.pcolormesh(ax=ax, vmin=230, vmax=300,
            x='lon', y='lat')
)
ax.coastlines();
```
